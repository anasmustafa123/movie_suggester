```python
import pandas as pd
import numpy as np
from  ast import literal_eval
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import ast
```

```python
shows = pd.read_csv("tvshows_df.csv")
```

```python
shows.head()
```

```python
def eachrowtolist(text):
  return ast.literal_eval(text)
```

```python
#convert all the rows to list instead of str to concatinate the them in the next step
shows["overview"] = shows["overview"].apply(eachrowtolist)
shows["tag"] = shows["tag"].apply(eachrowtolist)
```

```python
type(shows["overview"][0])
```

```python
# concatinate the tag with overview
shows["tag"]  = shows["tag"] + shows["overview"]
```

```python
#convert back to str
shows["tag"] = shows["tag"].apply(lambda x: f"[{', '.join(x)}]" if isinstance(x, list) else x)
shows["overview"] = shows["overview"].apply(lambda x: f"[{', '.join(x)}]" if isinstance(x, list) else x)
```

```python
#problem we have alot of english stopping points that dont contripute to the meaning
cv = CountVectorizer(max_features=shows.shape[0],stop_words='english')
```

```python
vector = cv.fit_transform(shows["tag"]).toarray()
```

```python
#applying similarity to the matrix
similarity = cosine_similarity(vector)
```

```python
def recommend(title):
  title_index =  shows[shows['title'] == title].index[0]
  shows_sorted = sorted(list(enumerate(similarity[title_index])), reverse=True, key = lambda x: x[1])
  for i in shows_sorted[0:15]:
    print(shows.iloc[i[0]].title)
```
